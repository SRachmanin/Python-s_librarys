```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
%config InlineBackend.figure_format = 'svg'
```

```python
x = [1, 2, 3, 4, 5, 6, 7]
y = [3.5, 3.8, 4.2, 4.5, 5.0, 5.5, 7.0]
```

```python
title_dict = {'fontsize': 11,
              'color': 'DimGrey',
              }
label_dict = {'fontsize': 9,
              'color': 'DarkGrey'}
```

```python
plt.plot(x, y, color = 'IndianRed')
plt.title('График y(x)', fontdict = title_dict)
plt.xlabel('Переменная x', fontdict = label_dict)
plt.ylabel('Функция y(x)', fontdict = label_dict)
plt.show()
```

```python
plt.scatter(x, y, color = 'IndianRed')
plt.title('График y(x)', fontdict = title_dict)
plt.xlabel('Переменная x', fontdict = label_dict)
plt.ylabel('Функция y(x)', fontdict = label_dict)
plt.show()
```

#### Задание 2

```python
t = np.linspace(0, 10, 51)
t
```

```python
f = np.cos(t)
f
```

```python
title_dict = {'fontsize': 11,
              'color': 'DimGrey',
              }
label_dict = {'fontsize': 9,
              'color': 'DarkGrey'}
```

```python
plt.axis([0.5, 9.5, -2.5, 2.5])
plt.plot(t, f, color = 'DarkOliveGreen')
plt.title('График f(t)', fontdict = title_dict)
plt.xlabel('Переменная t', fontdict = label_dict)
plt.ylabel('Функция f(t)', fontdict = label_dict)
plt.show()
```

#### Задание 3

```python
x = np.linspace(-3, 3, 51)
x
```

```python
y1 = x ** 2
y2 = 2* x + 0.5
y3 = -3 * x - 1.5
y4 = np.sin(x)
```

```python
title_dict = {'fontsize': 11,
              'color': 'DimGrey'}
```

```python
fig, ax = plt.subplots(nrows = 2, ncols = 2)
ax1, ax2, ax3, ax4 = ax.flatten()

ax1.plot(x, y1, color = 'LightBlue')
ax1.set_title('График y1', fontdict = title_dict)
ax1.set_xlim([-5, 5])

ax2.plot(x, y2, color = 'LightBlue')
ax2.set_title('График y2', fontdict = title_dict)

ax3.plot(x, y3, color = 'LightBlue')
ax3.set_title('График y3', fontdict = title_dict)

ax4.plot(x, y4, color = 'LightBlue')
ax4.set_title('График y4', fontdict = title_dict)

fig.set_size_inches(8, 6)
plt.subplots_adjust(wspace = 0.3, hspace = 0.3)
```

#### Задание 4

```python
import pandas as pd

plt.style.use('fivethirtyeight')

import warnings
warnings.filterwarnings('ignore')
```

```python
creditcard = pd.read_csv('creditcard.csv')
```

```python
class_list = creditcard['Class'].value_counts()
class_list
```

```python
class_list.plot(kind = 'barh', alpha=0.5, color = 'CadetBlue')
plt.show()
```

```python
class_list.plot(kind = 'barh', logx = True, alpha=0.5, color = 'Thistle')
plt.show()
```

```python
class0 = creditcard.loc[creditcard['Class'] == 0, ['V1']]
class1 = creditcard.loc[creditcard['Class'] == 1, ['V1']]
plt.hist(class0['V1'], bins=20, density=True, alpha=0.5, label='Class 0', color='grey')
plt.hist(class1['V1'], bins=20, density=True, alpha=0.5, label='Class 1', color='red')
plt.xlabel('V1', fontsize = 11)
plt.legend(fontsize = 7)
plt.show()
```
